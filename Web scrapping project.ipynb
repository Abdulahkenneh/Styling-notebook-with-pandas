```python
def openurl(url,href):
    try:   
        joined_url = urljoin(url,href)
        html = urlopen(joined_url)
        
        try:
            soup = BeautifulSoup(html,'html.parser')
            extract_content_from_table(soup)
            
        except AttributeError as e:
            print(e)
            
    except HTTPError as e:
        return e
```

```python
def extract_content_from_table(soup):
    headers = []
    data = dict()
    for i in soup.find_all('th'):
        if i not in headers:
            headers.append(i.get_text())
           
    list_of_td = [i.get_text() for i in soup.find_all('td')]
    print(list_of_td)
    for value in list_of_td:
            data.append(dict(zip(headers,value)))
            print('Done')

        

    print(data)
    print()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup

```

```python

# Set up the driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Open the eBay page
url = 'https://pixabay.com/videos/boy-talking-green-screen-cartoon-142558/'
driver.get(url)

# Get the page source after JavaScript has executed
html = driver.page_source

# Use BeautifulSoup to parse the HTML
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
# Print out the desired content
print(soup.prettify())

# Close the driver
driver.quit()
```
